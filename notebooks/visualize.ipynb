```python
import os
from matplotlib import pyplot as plt
import cv2
import json
```

```python
with open("../../datasets/dota_split/trainval_split/trainval.json", "r") as f:
    trainval = json.load(f)
```

```python
dirn = "../../datasets/dota_split/trainval_split/images"
fn = trainval[1]["filename"]
imgn = os.path.join(dirn, fn)
img = cv2.imread(imgn)
```

```python
import numpy as np
for row in trainval[1]["annotations"]['bboxes']:
    rect = ((row[0], row[1]), (row[2], row[3]), row[4]/np.pi*180)
    box = cv2.boxPoints(rect)
    box = np.int0(box)
    img = cv2.drawContours(img,[box],0,(0,0,255),2)
```

```python
plt.imshow(img)
```

```python
cv2.imwrite("output.png", img)
```
